```python
#write a program to plot a bar chart in python to display the result of a school for five consecutive years.
import matplotlib.pyplot as pl
year=['2010','2011','2012','2013','2014']
p=[98.50,70.25,55.20,90.5,61.50]
j=['b','g','r','m','c']
pl.bar(year,p,width=0.2,color=j)
pl.xlabel("year")
pl.ylabel("pass%")
pl.show()
```

```python
#write a program in python to plot a graph form the function y=x^2
import matplotlib.pyplot as plt
x_cords=range(-50,50)
y_cords=[x*x for x in x_cords]
plt.plot(x_cords,y_cords)
plt.show()
```

```python
#write a program in python to plot a pie chart chart on consumption of water in daily life
import matplotlib.pyplot as plt
A=['work','sleep','study','play']
B=[8,7,6,3]
plt.pie(B,labels=A,autopct='%1.1f%%')
plt.title('pie chart of daily activity')
plt.show()
```
