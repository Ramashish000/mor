```python
import numpy as np
import matplotlib.pyplot as plt

def linear_regression(x, y):
    """
    Perform linear regression and return the slope (m) and y-intercept (b)
    of the best-fit line (y = mx + b).
    """
    n = len(x)
    
    # Calculate the mean of x and y
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    
    # Calculate the slope (m) and y-intercept (b) of the best-fit line
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sum((x - mean_x) ** 2)
    
    m = numerator / denominator
    b = mean_y - m * mean_x
    
    return m, b

def plot_linear_regression(x, y, m, b):
    """Plot the original data points and the best-fit line."""
    plt.scatter(x, y, color='blue', label='Data points')
    plt.plot(x, m * x + b, color='red', label='Best-fit line')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Linear Regression')
    plt.legend()
    plt.show()

def main():
    # Sample data
    x = np.array([1, 2, 3, 4, 5])
    y = np.array([2, 3, 4, 5, 6])

    # Perform linear regression
    m, b = linear_regression(x, y)
    print(f"Slope (m): {m}")
    print(f"Y-intercept (b): {b}")

    # Plot the results
    plot_linear_regression(x, y, m, b)

if __name__ == "__main__":
    main()
```

```python
import csv

def write_to_csv(filename, data):
    """Write data to a CSV file."""
    with open(filename, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Name', 'Age', 'City'])  # Write header
        csv_writer.writerows(data)

def read_from_csv(filename):
    """Read data from a CSV file."""
    with open(filename, 'r') as csvfile:
        csv_reader = csv.reader(csvfile)
        header = next(csv_reader)  # Read header
        data = [row for row in csv_reader]
    return header, data

def main():
    # Sample data
    data_to_write = [
        ['Alice', 25, 'New York'],
        ['Bob', 30, 'San Francisco'],
        ['Charlie', 22, 'Chicago']
    ]

    # Specify the CSV file name
    csv_filename = 'sample_data.csv'

    # Write data to CSV
    write_to_csv(csv_filename, data_to_write)
    print(f'Data written to {csv_filename}')

    # Read data from CSV
    header, data_read = read_from_csv(csv_filename)
    print(f'\nHeader: {header}')
    print('Data read from CSV:')
    for row in data_read:
        print(row)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def enter_data():
    """Allow the user to enter data for multiple columns and rows."""
    data = {}

    num_columns = int(input("Enter the number of columns: "))
    column_names = [input(f"Enter the name for column {i + 1}: ") for i in range(num_columns)]

    num_rows = int(input("Enter the number of rows: "))

    for column in column_names:
        data[column] = [input(f"Enter data for {column} in row {i + 1}: ") for i in range(num_rows)]

    return data

def display_data(data):
    """Display the entered data using a pandas DataFrame."""
    df = pd.DataFrame(data)
    print("\nData entered:\n", df)

def main():
    data = enter_data()
    display_data(data)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def generate_sample_data():
    """Generate a sample dataset for demonstration."""
    data = {
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'Age': [25, 30, 22, 35, 28],
        'Salary': [50000, 60000, 45000, 70000, 55000]
    }
    return pd.DataFrame(data)

def perform_statistical_measures(df):
    """Perform various statistical measures on the dataset."""
    print("Summary Statistics:")
    print(df.describe())

    print("\nCorrelation Matrix:")
    print(df.corr())

    print("\nCovariance Matrix:")
    print(df.cov())

    print("\nUnique Values in 'Name' column:")
    print(df['Name'].unique())

    print("\nAverage Age:", df['Age'].mean())
    print("Maximum Salary:", df['Salary'].max())
    print("Minimum Age:", df['Age'].min())

def main():
    # Generate a sample dataset
    df = generate_sample_data()

    # Display the dataset
    print("Sample Dataset:")
    print(df)

    # Perform statistical measures
    perform_statistical_measures(df)

if __name__ == "__main__":
    main()
```

```python
def calculate_grade(total_marks):
    """Calculate and return the grade based on total marks."""
    if total_marks >= 90:
        return 'A'
    elif 80 <= total_marks < 90:
        return 'B'
    elif 70 <= total_marks < 80:
        return 'C'
    elif 60 <= total_marks < 70:
        return 'D'
    elif 50 <= total_marks < 60:
        return 'E'
    else:
        return 'F'

def main():
    print("Enter marks for 5 subjects:")

    marks = []
    for i in range(5):
        subject_marks = float(input(f"Enter marks for subject {i + 1}: "))
        marks.append(subject_marks)

    # Calculate the total marks
    total_marks = sum(marks)

    # Calculate the grade
    grade = calculate_grade(total_marks)

    # Display the result
    print(f"\nTotal Marks: {total_marks}")
    print(f"Grade: {grade}")

if __name__ == "__main__":
    main()
```
