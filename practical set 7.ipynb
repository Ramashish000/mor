```python
import matplotlib.pyplot as plt

def plot_bar_chart(years, results):
    """Plot a bar chart for school results over five consecutive years."""
    plt.bar(years, results, color='skyblue')
    plt.xlabel('Year')
    plt.ylabel('Result')
    plt.title('School Results Over Five Years')
    plt.show()

def main():
    # Sample data: Results for five consecutive years
    years = ['Year 1', 'Year 2', 'Year 3', 'Year 4', 'Year 5']
    results = [85, 90, 88, 92, 95]

    # Plot the bar chart
    plot_bar_chart(years, results)

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_function():
    """Plot the graph for the function y = x^2."""
    # Generate x values
    x = np.linspace(-10, 10, 100)  # You can adjust the range and number of points

    # Calculate corresponding y values
    y = x**2

    # Plot the graph
    plt.plot(x, y, label='y = x^2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Graph of y = x^2')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    # Plot the graph for y = x^2
    plot_function()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

def plot_water_consumption():
    """Plot a pie chart for water consumption in daily life."""
    categories = ['Drinking', 'Cooking', 'Bathing', 'Cleaning', 'Other']
    consumption = [40, 10, 25, 15, 10]  # Sample data in percentage

    # Plot the pie chart
    plt.pie(consumption, labels=categories, autopct='%1.1f%%', startangle=90)
    plt.title('Water Consumption in Daily Life')
    plt.show()

def main():
    # Plot the pie chart for water consumption
    plot_water_consumption()

if __name__ == "__main__":
    main()
```
